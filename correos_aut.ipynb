## ✉️ **Automatizador de Envío de Correos a Transportistas con DETALLE, Anticipos y Saldos**

✅ Genera con consultas las **PLANILLAS**: ***Anticipo o Saldo / Líquidos o Seco***

```python
#TODOEN UNO CON CONSULTAS 

import pandas as pd

# === Input interactivo ===

print("=== FILTRADO DE DATOS DE EXCEL ===")

# Modo de operación: ANTICIPO o SALDO
modo = input("¿Deseas trabajar con ANTICIPO o SALDO? ").strip().upper()

# Tipo de producto: SECO o LÍQUIDOS (se normaliza a LIQUIDOS si no tiene tilde)
tipo_producto = input("¿El producto es SECO o LÍQUIDOS? ").strip().upper().replace("Í", "I")

# === Validación de entradas ===

if modo not in ["ANTICIPO", "SALDO"]:
    raise ValueError("❌ Opción no válida para MODO. Usa 'ANTICIPO' o 'SALDO'.")

if tipo_producto not in ["SECO", "LIQUIDOS"]:
    raise ValueError("❌ Opción no válida para el tipo de producto. Usa 'SECO' o 'LÍQUIDOS'.")

# === Configuración de hoja ===

if tipo_producto == "SECO":
    hoja = "Detalles (Seco)"
else:
    hoja = "Detalles (Liquido)"

# === Configuración según modo y tipo de producto ===

archivo_excel = 'Plantilla.xlsx'

if modo == "ANTICIPO" and tipo_producto == "SECO":
    hoja = "Detalles (Seco)"
    columna_fecha = 'FECHA PAGO ANTICIPO'
    columnas_deseadas = [
        'TRANSPORTADORA',
        'PLACA',
        'CHOFER',
        'LICENCIA',
        'FACTURA',
        'BOOKING',
        'PESO PROGR. TM',
        'PRODUCTO',
        'FLETE NEGOCIADO2',
        'FLETE NEGOCIADO',
        'MONTO ANTICIPO',
        'FECHA PAGO ANTICIPO',
        'BANCO ANTICIPO',
        'SALDO FLETE A CANCELAR',
    ]
    archivo_salida = 'datos_filtrados_anticipo_seco.xlsx'

elif modo == "SALDO" and tipo_producto == "SECO":
    hoja = "Detalles (Seco)"
    columna_fecha = 'FECHA PAGO SALDO FLETE'
    columnas_deseadas = [
        'TRANSPORTADORA',
        'PLACA',
        'CHOFER',
        'LICENCIA',
        'FACTURA',
        'BOOKING',
        'PESO PROGR. TM',
        'PESO NETO DESCARGADO',
        'PRODUCTO',
        'FLETE NEGOCIADO2',
        'FLETE NEGOCIADO',    
        'SALDO FLETE A CANCELAR',
        'FECHA PAGO SALDO FLETE',
        'BANCO SALDO',
        
    ]
    archivo_salida = 'datos_filtrados_saldo_seco.xlsx'

elif modo == "ANTICIPO" and tipo_producto == "LIQUIDOS":
    hoja = "Detalles (Liquido)"
    columna_fecha = 'FECHA PAGO ANTICIPO'
    columnas_deseadas = [
        'TRANSPORTADORA',
        'PLACA',
        'CHOFER',
        'LICENCIA',
        'PESO PROGRAMADO (TM)',
        'PRODUCTO',
        'FACTURA',
        'BOOKING;FACTURA NUTREX',
        'FLETE NEGOCIADO',
        'ANTICIPO',
        'FECHA PAGO ANTICIPO',
        'BANCO ANTICIPO',
        'OBS',
    ]
    archivo_salida = 'datos_filtrados_anticipo_liquido.xlsx'

elif modo == "SALDO" and tipo_producto == "LIQUIDOS":
    hoja = "Detalles (Liquido)"
    columna_fecha = 'FECHA SALDO'
    columnas_deseadas = [
        'TRANSPORTADORA',
        'PLACA',
        'CHOFER',
        'LICENCIA',
        'PESO PROGRAMADO (TM)',
        'PESO NETO DESCARGADO',
        'PRODUCTO',
        'FACTURA',
        'BOOKING;FACTURA NUTREX',
        'FLETE NEGOCIADO',
        'SALDO',
        'FECHA SALDO',
        'BANCO SALDO',
        'OBS',
    ]
    archivo_salida = 'datos_filtrados_saldo_liquido.xlsx'

else:
    raise ValueError("Combinación de MODO y PRODUCTO no válida.")


# === Confirmación de configuración ===
print("\n--- Configuración seleccionada ---")
print(f"Modo: {modo}")
print(f"Producto: {tipo_producto}")
print(f"Hoja de Excel a usar: {hoja}")
print(f"Columna de fecha: {columna_fecha}")

# === Procesamiento ===

try:
    # Leer la hoja correspondiente
    df = pd.read_excel(archivo_excel, sheet_name=hoja)

    # Convertir la columna de fecha a datetime
    df[columna_fecha] = pd.to_datetime(df[columna_fecha], errors='coerce')

    # Obtener la fecha más reciente válida
    fecha_mas_reciente = df[columna_fecha].max().date()

    # Filtrar por esa fecha
    df_filtrado = df[df[columna_fecha].dt.date == fecha_mas_reciente]

    # Verificar columnas necesarias
    faltantes = [col for col in columnas_deseadas if col not in df_filtrado.columns]
    if faltantes:
        raise ValueError(f"❌ Las siguientes columnas no existen en la hoja seleccionada: {faltantes}")

    # Seleccionar solo las columnas deseadas
    df_filtrado = df_filtrado[columnas_deseadas]

    # Guardar el archivo de salida
    df_filtrado.to_excel(archivo_salida, index=False)

    # Éxito
    print("✅ PROCESO COMPLETADO")
    print(f"Archivo generado: {archivo_salida}")
    print(f"Fecha usada para filtrar: {fecha_mas_reciente}")

except Exception as e:
    print("\n❌ ERROR durante la ejecución:")
    print(str(e))
    
```

```python
import pandas as pd
import win32com.client as win32
import os

# === Configuración ===
archivo_excel = 'datos_filtrados_anticipo_seco.xlsx'
archivo_remitentes = 'remitentes.xlsx'
columna_transportadora = 'TRANSPORTADORA'
columna_para = 'PARA'
columna_cc = 'CC'

# Ruta donde se encuentran los comprobantes PDF
ruta_adjuntos = r'C:\Users\Nutrex\OneDrive - NUTREX PARAGUAY SRL\Pago_transportistas\COMPROBANTES'

# Leer archivos
df = pd.read_excel(archivo_excel)
remitentes_df = pd.read_excel(archivo_remitentes)


# Estandarizar nombres para asegurar coincidencia
df[columna_transportadora] = df[columna_transportadora].str.strip().str.upper()
remitentes_df[columna_transportadora] = remitentes_df[columna_transportadora].str.strip().str.upper()

# Unir los datos con los correos
df = df.merge(remitentes_df, on=columna_transportadora, how='left')

# Verificar si faltan correos
faltantes = df[df[columna_para].isna()][columna_transportadora].unique()
if len(faltantes) > 0:
    print("❌ Las siguientes transportadoras no tienen correo en columna 'PARA':")
    for f in faltantes:
        print(f" - {f}")
    raise ValueError("Agrega estos correos en el archivo de remitentes.")

# Iniciar Outlook
outlook = win32.Dispatch('Outlook.Application')

# Estilo CSS
estilo_tabla = """
<style>
    table {
        border-collapse: collapse;
        width: 100%;
        font-family: Arial, sans-serif;
        font-size: 12px;
    }
    th {
        background-color: #4A90E2;
        color: white;
        font-weight: bold;
        padding: 8px;
        border: 1px solid #ddd;
        text-align: center;
    }
    td {
        padding: 8px;
        border: 1px solid #ddd;
        text-align: center;
    }
    tr:nth-child(even) {
        background-color: #f2f2f2;
    }
</style>
"""

# Agrupar por transportadora
for transportadora, grupo in df.groupby(columna_transportadora):
    correo_para = grupo[columna_para].iloc[0]
    correo_cc = grupo[columna_cc].iloc[0] if columna_cc in grupo.columns else ''

    # Convertir nombre de transportadora a mayúsculas para buscar en nombre de archivo
    nombre_busqueda = transportadora.upper()

    # Buscar archivo PDF que contenga el nombre del transportista
    adjunto_encontrado = None
    for archivo in os.listdir(ruta_adjuntos):
        if archivo.lower().endswith('.pdf') and nombre_busqueda in archivo.upper():
            adjunto_encontrado = os.path.join(ruta_adjuntos, archivo)
            break

    # Generar la tabla HTML
    tabla_html = grupo.drop(columns=[columna_para, columna_cc]).to_html(index=False, border=0, justify='center', escape=False)

    cuerpo_html = f"""
    <html>
    <head>{estilo_tabla}</head>
    <body>
        <p>Buenos días,</p>
        <p>En adjunto el comprobante y el detalle de pago de Anticipo Seco realizado.</p>

        {tabla_html}

        <p>Por favor si pueden enviar un recibo como comprobante de que el pago fue recibido, ¡muchas gracias!</p>
        <p>Tener en cuenta que no recibiremos reclamos posteriores a los dos meses de emisión de este comprobante.</p>
        <p>Saludos cordiales,<br></p>

        <span>Natalye Silva - Verificaciones,</span><br>
        <span>NUTREX INC.</span><br>
        <span>verificacion@nutrexinc.com</span>
    </body>
    </html>
    """


    # ... (aquí terminas de leer y unir los datos con remitentes)

    # === para definir fecha mas reciente si no se corre el script anterior primero ===
    candidatas = ['FECHA PAGO ANTICIPO', 'FECHA PAGO SALDO FLETE', 'FECHA SALDO']
    col_fecha = next((c for c in candidatas if c in df.columns), None)
    if col_fecha is None:
        posibles = [c for c in df.columns if 'FECHA' in c.upper()]
        from datetime import date
        fecha_mas_reciente = pd.Timestamp(date.today()) if not posibles else pd.to_datetime(df[posibles[0]], errors='coerce').max()
    else:
        df[col_fecha] = pd.to_datetime(df[col_fecha], errors='coerce')
        fecha_mas_reciente = df[col_fecha].max()

   
    # Crear el correo
    mail = outlook.CreateItem(0)
    mail.To = correo_para
    if correo_cc:
        mail.CC = correo_cc
    mail.Subject = f"PAGO TRANSPORTISTA {fecha_mas_reciente.strftime('%d/%m/%Y')} - ANTICIPO - {transportadora}"
    mail.HTMLBody = cuerpo_html
    
    # Si se encontró un adjunto, lo agrega
    if adjunto_encontrado and os.path.exists(adjunto_encontrado):
        mail.Attachments.Add(adjunto_encontrado)
        print(f"📎 Adjunto agregado para {transportadora}: {os.path.basename(adjunto_encontrado)}")
    else:
        print(f"⚠️ No se encontró adjunto para {transportadora} (buscando por: '{nombre_busqueda}')")

    # Guardar en borradores
    mail.Save()

print("\n✅ Todos los correos fueron guardados en Borradores con destinatarios correctos.")


# ANTICIPO SECO 
```
